```
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv")
data.drop(["Channel", "Region"], axis = 1, inplace = True)
```

```
data.head()
```

Groceries: The customer’s annual spending on grocery products.

Milk: The customer’s annual spending on milk products.

```
fdata = data[["Grocery", "Milk"]]
data = fdata.to_numpy().astype("float32", copy = False)
```

```
fdata.plot.scatter(x="Grocery",y="Milk",c="blue")
```

normalize each attribute by scaling it to 0 mean and unit variance

```
stscaler = StandardScaler().fit(data)
data = stscaler.transform(data)
```

```
dbsc = DBSCAN(eps = .5, min_samples = 15).fit(data)
labels = dbsc.labels_
core_samples = np.zeros_like(labels, dtype = bool)
core_samples[dbsc.core_sample_indices_] = True
```
